```python
def room_filter(s):
    if '合住房間' in s:
        return '合住房間'
    elif '獨立房間' in s :
        return '獨立房間'
    elif '房間' in s:
        return '房間'
    else :
        return '整個房源'
df['room_type']=df['room_and_property_type'].apply(room_filter)
dummy = pd.get_dummies(df['room_type'])
df = pd.concat([df, dummy], axis=1)
```

```python
df.columns
```

```python
amenitys=set()
for s in df['preview_amenity_names']:
    for i in s.replace("'","").replace('[','').replace(']','').replace(' ','').split(','):
        amenitys.add(i)
        
for amenity in amenitys:
    df[amenity]=0
    for i,s in enumerate(df['preview_amenity_names']):
        df.loc[i,[amenity]]=1 if amenity in s else 0
```

```python
df1=df[['displayPrice','rating', '合住房間', '房間', '整個房源', '獨立房間']]#,'無線網路','空調設備','廚房','洗衣機'
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
scaler = MinMaxScaler()
cluster=AgglomerativeClustering(n_clusters=5)#KMeans(n_clusters=10)
df1=pd.DataFrame(scaler.fit_transform(df1),columns=df1.columns)
df1['result']=cluster.fit_predict(df1)
```

```python
plt.scatter(x=df1['displayPrice'],y=df1['rating'],c=df1['獨立房間'])
```
